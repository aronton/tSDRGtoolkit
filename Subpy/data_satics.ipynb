```python
import os
import math
import time
import timeit
import numpy as np
import sys
import tarfile
import datetime
import multiprocessing
import scriptCreator
from pathlib import Path
import shutil


base_random = "/home/aronton/tSDRG_random/tSDRG/Main_15/data_random"
base_collect = "/home/aronton/tSDRG_random/tSDRG/Main_15/data_collect"
base_metadata = "/home/aronton/tSDRG_random/tSDRG/Main_15/metadata"
base_old = "/homeold/aronton/tSDRG_random/tSDRG/Main_15/data_random"
P = 10
BC = "PBC"
chi = 40

Jstr = [f"Jdis{str(i).zfill(3)}" for i in range(10,201,10)]
Dstr = [f"Dim{str(i).zfill(3)}" for i in range(1)]
Lstr = [f"L{num}" for num in range(64, 550, 64)]  # 只有 L512

print(f"original data on scopion:\n {base_random}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            for s,seed in  enumerate(list(range(10000,1000,-1000))):
                # output += f"{base_random}/{BC}/{J}/{D}/{L}_P{P}_m{chi}_{seed}  "
                # print(output)
                if os.path.exists(f"{base_random}/{BC}/{J}/{D}/{L}_P{P}_m{chi}_{seed}"):
                    output += f"{L}:{seed}, "
                    # output += f"{L}_P{P}_m{chi}_{seed}  "
                    break
            output += f"{L}:0, "
        print(f"{J}:{output}")

print(f"collect data on scopion:\n {base_collect}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            path = os.path.join(base_collect, BC, J, D, f"{L}_P{P}_m{chi}")
            filename = f"ZL_{L}_P{P}_m{chi}_{J}_{D}.txt"
            filepath = os.path.join(path, filename)

            try:
                with open(filepath, 'r') as file:
                    lines = file.readlines()
                length = len(lines)-1
                output += f"{L}:{length}, "
            except FileNotFoundError:
                continue
            except Exception as e:
                continue
                # output += f"{L}_P{P}_m{chi}_{seed}  "
                
        print(f"{J}:{output}")
        
print(f"metadata data on scopion:\n {base_metadata}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            path = os.path.join(base_metadata, BC, J, D, f"{L}_P{P}_m{chi}_meta")
            filename = f"ZL_{L}_P{P}_m{chi}_{J}_{D}.txt"
            filepath = os.path.join(path, filename)

            try:
                with open(filepath, 'r') as file:
                    lines = file.readlines()
                del lines[0]  # 删除第一行
                length = lines[0].split(",")[1]
                output += f"{L}:{length}, "
            except FileNotFoundError:
                continue
            except Exception as e:
                continue
                # output += f"{L}_P{P}_m{chi}_{seed}  "
                
        print(f"{J}:{output}")
        
```
