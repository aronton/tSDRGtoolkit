```python
#!/usr/bin/env python3
import os
import tarfile

# === 你可以調整這個參數 ===
BASE_DIR = "/home/aronton/tSDRG_random/tSDRG/Main_15/tarNew/OBC"   # 包含 Jdis001 ~ Jdis200 的上層資料夾
J_START, J_END = 1, 201          # Jdis 範圍
D_PREFIX = "Dim"                 # tar 檔案名稱前綴

def extract_tar(tar_path, dest_dir):
    """解壓 tar 檔案到指定資料夾"""
    try:
        print(f"解壓中: {tar_path}")
        with tarfile.open(tar_path, "r") as tar:
            tar.extractall(path=dest_dir)
    except Exception as e:
        print(f"❌ 解壓失敗 {tar_path}: {e}")

def process_j_folder(j_folder):
    """處理單個 Jdis 資料夾"""
    print(f"=== 處理 {j_folder} ===")
    tar_files = [f for f in os.listdir(j_folder) if f.endswith(".tar")]

    if not tar_files:
        print(f"⚠️ {j_folder} 沒有 tar 檔案，跳過")
        return

    for tar_file in tar_files:
        tar_path = os.path.join(j_folder, tar_file)
        extract_tar(tar_path, j_folder)

def main():
    # 依序檢查 Jdis001 ~ Jdis200
    for j in range(J_START, J_END + 1):
        j_name = f"Jdis{j:03d}"
        j_path = os.path.join(BASE_DIR, j_name)

        if not os.path.isdir(j_path):
            print(f"❌ 找不到資料夾 {j_name}，跳過")
            continue

        process_j_folder(j_path)

if __name__ == "__main__":
    main()
```

```python
import os
import math
import time
import timeit
import numpy as np
import sys
import tarfile
import datetime
import multiprocessing
import scriptCreator
from pathlib import Path
import shutil


base_random = "/home/aronton/tSDRG_random/tSDRG/Main_15/data_random"
base_collect = "/home/aronton/tSDRG_random/tSDRG/Main_15/data_collect"
base_metadata = "/home/aronton/tSDRG_random/tSDRG/Main_15/metadata"
base_old = "/homeold/aronton/tSDRG_random/tSDRG/Main_15/data_random"
P = 10
BC = "PBC"
chi = 40

Jstr = [f"Jdis{str(i).zfill(3)}" for i in range(10,201,10)]
Dstr = [f"Dim{str(i).zfill(3)}" for i in range(1)]
Lstr = [f"L{num}" for num in range(64, 550, 64)]  # 只有 L512

print(f"\n\noriginal data on scopion:\n {base_random}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            for s,seed in  enumerate(list(range(10000,1000,-1000))):
                # output += f"{base_random}/{BC}/{J}/{D}/{L}_P{P}_m{chi}_{seed}  "
                # print(output)
                if os.path.exists(f"{base_random}/{BC}/{J}/{D}/{L}_P{P}_m{chi}_{seed}"):
                    output += f"{L}:{int(seed):>5}, "
                    # output += f"{L}_P{P}_m{chi}_{seed}  "
                    break
            output += f"{L}:0, "
        print(f"{J}:{output}")

print(f"\n\ncollect data on scopion:\n {base_collect}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            path = os.path.join(base_collect, BC, J, D, f"{L}_P{P}_m{chi}")
            filename = f"ZL_{L}_P{P}_m{chi}_{J}_{D}.txt"
            filepath = os.path.join(path, filename)

            try:
                with open(filepath, 'r') as file:
                    lines = file.readlines()
                length = len(lines)-1
                output += f"{L}:{int(length):>5}, "
            except FileNotFoundError:
                continue
            except Exception as e:
                continue
                # output += f"{L}_P{P}_m{chi}_{seed}  "
                
        print(f"{J}:{output}")
        
print(f"\n\nmetadata data on scopion:\n {base_metadata}\n")
for j,J in enumerate(Jstr):
    for d,D in enumerate(Dstr):
        # output += f"{D}:"
        output = ""
        for l,L in enumerate(Lstr):
            path = os.path.join(base_metadata, BC, J, D, f"{L}_P{P}_m{chi}_meta")
            filename = f"ZL_{L}_P{P}_m{chi}_{J}_{D}.txt"
            filepath = os.path.join(path, filename)

            try:
                with open(filepath, 'r') as file:
                    lines = file.readlines()
                del lines[0]  # 删除第一行
                length = lines[0].split(",")[1]
                output += f"{L}:{int(length):>5}, "
            except FileNotFoundError:
                continue
            except Exception as e:
                continue
                # output += f"{L}_P{P}_m{chi}_{seed}  "
                
        print(f"{J}:{output}")
        
```
