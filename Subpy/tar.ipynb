```python
import os
import tarfile
from multiprocessing import Pool, cpu_count

def make_tar(task):
    origin_dir, tar_dir = task
    if not os.path.exists(origin_dir):
        print(f"[Skip] Source not found: {origin_dir}")
        return
    print(f"origin_dir :{origin_dir}")
    os.makedirs(tar_dir, exist_ok=True)
    for item in os.listdir(origin_dir):
        item_path = os.path.join(origin_dir, item)
        if os.path.isdir(item_path):
            tar_file = os.path.join(tar_dir, f"{item}.tar")
            if os.path.exists(tar_file):
                print(f"[Skip] Already exists: {tar_file}")
                continue
            try:
                with tarfile.open(tar_file, "w") as tar:
                    tar.add(item_path, arcname=os.path.basename(item_path))
                print(f"[OK] Packed: {item_path} -> {tar_file}")
            except Exception as e:
                print(f"[Error] Failed to pack {item_path}: {e}")

def build_task_list(base_src, base_tar, BC, J_list):
    tasks = []
    for J in J_list:
        src_path = os.path.join(base_src, BC, J)
        tar_path = os.path.join(base_tar, BC, J)
        tasks.append((src_path, tar_path))
    return tasks

if __name__ == "__main__":
    # 設定參數
    BC = "OBC"  # 或 "OBC"
    J_list = [f"Jdis{str(i).zfill(3)}" for i in range(1, 201, 1)]
    # D_list = [f"Dim{str(i).zfill(3)}" for i in range(10, 12, 5)]

    base_src = "/homeold/aronton/tSDRG_random/tSDRG/Main_15/data_random"
    base_tar = "/home/aronton/tSDRG_random/tSDRG/Main_15/tar"

    # 建立任務清單
    task_list = build_task_list(base_src, base_tar, BC, J_list)

    # 啟動 multiprocessing
    workers = min(cpu_count()/2, 12)  # 可依需求限制核心數，例如最多 12 核
    with Pool(processes=workers) as pool:
        pool.map(make_tar, task_list)
```
